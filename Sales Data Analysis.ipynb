```python
#importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#printing the total sales
total_sales = df['Sales'].sum()
print(f'Total Sales: ${total_sales:,.2f}')
```

```python
#analyzing sales trends over time
monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
#finding the best selling products
product_sales = df.groupby('Product Name')['Sales'].sum().reset_index()
product_sales = product_sales.sort_values(by='Sales', ascending=False)

print(product_sales.head())
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sales', y='Product Name', data=product_sales.head(10))
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```
